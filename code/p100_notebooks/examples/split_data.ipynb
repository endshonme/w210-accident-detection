```python
import os
collision_list = os.listdir('/workspace/data/training_data2/collision')
no_collision_list = os.listdir('/workspace/data/training_data2/no_collision')
```

```python
import random
import numpy as np
train_pct = 0.8
val_pct = 0.5
test_pct = 0.5
```

```python
random.seed(3)
train_collision = random.sample(collision_list, np.round(train_pct*len(collision_list)).astype(int))
train_no_collision = random.sample(no_collision_list, np.round(train_pct*len(no_collision_list)).astype(int))
```

```python
res_collision_list = set(collision_list) - set(train_collision)
res_no_collision_list = set(no_collision_list) - set(train_no_collision)
```

```python
random.seed(3)
val_collision = random.sample(res_collision_list, np.round(val_pct*len(res_collision_list)).astype(int))
val_no_collision = random.sample(res_no_collision_list, np.round(val_pct*len(res_no_collision_list)).astype(int))
```

```python
test_collision = set(res_collision_list) - set(val_collision)
test_no_collision = set(res_no_collision_list) - set(val_no_collision)
```

```python
print(" Test set [collision]= {}".format(len(test_collision)))
print(" Test set [no_collision]= {}".format(len(test_no_collision)))
print("\t---")
print(" val set [collision]= {}".format(len(val_collision)))
print(" val set [no_collision]= {}".format(len(val_no_collision)))
print("\t---")
print(" Train set [collision]= {}".format(len(train_collision)))
print(" Train set [no_collision]= {}".format(len(train_no_collision)))
```

### ANOTHER APPROACH TO FOCUS ON GENERALIZATION

We might want to evaluate the splitting could influence the ability of the model to generalize.

```python
collision_list_vid = [f.split('_')[0] for f in collision_list]
no_collision_list_vid = [f.split('_')[0] for f in no_collision_list]
```

```python
#Find the unique videos to divide into train, validate and test
prefix_vid = []
for f in no_collision_list:
    prefix_vid.append(f.split('_')[0])
for f in collision_list:
    prefix_vid.append(f.split('_')[0])
unique_video_id = set(prefix_vid)
total_videos = len(unique_video_id)
print('Total videos: {}'.format(total_videos))
```

```python
random.seed(3)
train_vid = random.sample(unique_video_id, np.round(train_pct*total_videos).astype(int))
res_video_id = list(set(unique_video_id) - set(train_vid))
val_vid = random.sample(res_video_id, np.round(val_pct*len(res_video_id)).astype(int))
test_vid = list(set(res_video_id) - set(val_vid))
print('Train:{} Validate:{} Test:{}'.format(len(train_vid),len(val_vid),len(test_vid)))
```
