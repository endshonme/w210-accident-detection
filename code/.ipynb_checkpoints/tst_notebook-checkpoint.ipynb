```python
def add_labels(frame_loc = '/home/saifrais/w251/final_project/data/video_sample_test/frames_1fps/vid_0031.jpg',
               labels_loc = '/tmp/w251-accident-detection/bin/performance_1fps.csv',
               output_path = '/home/saifrais/w251/final_project/data/video_sample_test/results_1fps'):
    import pandas as pd
    from PIL import Image
    from PIL import ImageDraw 
    import os
    
    perf = pd.read_csv(labels_loc)
    df = perf[perf['url']==frame_loc]
    status_str = df.label_pred.values.astype(str)[0].upper()
    status_prob = df.prob.round(2).values.astype(str)[0]
    alert_msg = status_str+':'+status_prob

    
    img = Image.open(frame_loc)
    draw = ImageDraw.Draw(img)
    #draw.text((250, 170),alert_msg,(255, 0, 0))
    if status_str=='COLLISION':# and df.prob.round(2).values > 0.75:
        draw.text((250, 170),alert_msg,(255, 0, 0))
    else:
        draw.text((250, 170),alert_msg,(255, 255, 255))
    
    frame_name = os.path.split(frame_loc)[1]
    res_loc = os.path.join(result_path,frame_name)
    img.save(res_loc)
    print('Image stored')

add_labels(frame_loc = '/home/saifrais/w251/final_project/data/video_sample_test/frames_1fps/vid_0030.jpg',
          labels_loc = '/tmp/w251-accident-detection/bin/performance_1fps.csv',
          output_path = '/home/saifrais/w251/final_project/data/video_sample_test/results_1fps')    
```

```python
frame_path = '/home/saifrais/w251/final_project/data/video_sample_test/frames_1fps'
labels_loc = '/tmp/w251-accident-detection/bin/performance_1fps.csv'
output_path = '/home/saifrais/w251/final_project/data/video_sample_test/results_1fps'
count = 0
for frame_name in video_1fps:
    frame_loc = os.path.join(frame_path,frame_name)
    print(frame_loc)
    add_labels(frame_loc = frame_loc, labels_loc = labels_loc, output_path = output_path)
    count = count + 1
    print('Frame no. {}'.format(count))
```
