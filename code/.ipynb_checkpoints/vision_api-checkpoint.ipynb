```python
#https://github.com/GoogleCloudPlatform/python-docs-samples/blob/master/storage/cloud-client/snippets.py
# Accessing the Google Cloud Storage
from google.cloud import storage

def list_blobs(bucket_name):
    """Lists all the blobs in the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)

    blobs = bucket.list_blobs()

    for blob in blobs:
        print(blob.name)
        
def list_blobs_with_prefix(bucket_name, prefix, delimiter=None):
    """Lists all the blobs in the bucket that begin with the prefix.

    This can be used to list all blobs in a "folder", e.g. "public/".

    The delimiter argument can be used to restrict the results to only the
    "files" in the given "folder". Without the delimiter, the entire tree under
    the prefix is returned. For example, given these blobs:

        /a/1.txt
        /a/b/2.txt

    If you just specify prefix = '/a', you'll get back:

        /a/1.txt
        /a/b/2.txt

    However, if you specify prefix='/a' and delimiter='/', you'll get back:

        /a/1.txt

    """
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)

    blobs = bucket.list_blobs(prefix=prefix, delimiter=delimiter)
    
    #print('Blobs:')
    #for blob in blobs:
    #    print(blob.name)

    #if delimiter:
    #    print('Prefixes:')
    #    for prefix in blobs.prefixes:
    #        print(prefix)
    
    all_img_uri = []
    for img_uri in blobs:
        all_img_uri.append(img_uri.name)
        len(all_img_uri)
            
    return(all_img_uri)

# [START storage_upload_file]
def upload_blob(bucket_name, source_file_name, destination_blob_name):
    """Uploads a file to the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)

    blob.upload_from_filename(source_file_name)

    print('File {} uploaded to {}.'.format(
        source_file_name,
        destination_blob_name))
# [END storage_upload_file]

def download_blob(bucket_name, source_blob_name, destination_file_name):
    """Downloads a blob from the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blob = bucket.blob(source_blob_name)

    blob.download_to_filename(destination_file_name)

    print('Blob {} downloaded to {}.'.format(
        source_blob_name,
        destination_file_name))


def delete_blob(bucket_name, blob_name):
    """Deletes a blob from the bucket."""
    storage_client = storage.Client()
    bucket = storage_client.get_bucket(bucket_name)
    blob = bucket.blob(blob_name)

    blob.delete()

    print('Blob {} deleted.'.format(blob_name))
```

```python
import io
import os
# Imports the Google Cloud client library
from google.cloud import vision
from google.cloud.vision import types

def detect_labels(vision_client, img_uri, dict_value = True):
    #client = vision.ImageAnnotatorClient()
    gcs_img = types.ImageSource(gcs_image_uri = img_uri)
    image = types.Image(source=gcs_img)
    response = client.label_detection(image=image, max_results = 100)
    labels = response.label_annotations
    
    if dict_value == True:
        lbl_dict = dict()
        for lbl in labels:
            lbl_dict[lbl.description] = lbl.topicality
        return lbl_dict
    else:
        import pandas as pd
        col_names =  ['mid', 'description', 'score','topicality']
        df  = pd.DataFrame(columns = col_names)
        for lbl in labels:
            tmp_df = pd.DataFrame({'mid':[lbl.mid],'description':[lbl.description],
                                   'score':[lbl.score],'topicality':[lbl.topicality]})
            df = df.append(tmp_df,ignore_index = True)
        return df

# Function to calculate the presence of a label
def check_description(lbl_dict,tgt_lbl = 'Crash'):
    return tgt_lbl in list(lbl_dict.keys())
```

```python
#Use pickle to store results
#https://pythonprogramming.net/python-pickle-module-save-objects-serialization/
import pickle
def save_obj(obj, storeloc = "dict.pickle"):
    pickle_out = open(storeloc,"wb")
    pickle.dump(obj, pickle_out)
    print('{} saved'.format(storeloc))
    pickle_out.close()
    
def load_obj(storeloc="accident_df.pickle"):
    import pandas as pd
    pickle_in = open(storeloc,"rb")
    obj = pickle.load(pickle_in)
    print('{} loaded'.format(storeloc))
    return obj
```

```python
import pandas as pd
col_names = ['gcs_uri','accident']
accident_df = pd.DataFrame(columns = col_names)
#tmp_df = pd.DataFrame(columns =col_names)
all_dict_lbls = dict()
tmp_dict = dict()
#Save the information every 100 images
#for uri in all_crash_uri:
all_idx = np.arange(len(all_crash_uri))
for idx in np.arange(len(all_crash_uri)):
    uri = all_crash_uri[idx]
    full_uri = os.path.join('gs://',bucket_name,uri)
    tmp_dict = detect_labels(client,full_uri, dict_value = True)
    accident_flag = any([check_description(tmp_dict,'Crash'),
                         check_description(tmp_dict,'Event'),
                        check_description(tmp_dict,'Glass')])
    tmp_df = pd.DataFrame({'gcs_uri':[full_uri],'accident':[accident_flag]})
    all_dict_lbls[full_uri] = tmp_dict
    accident_df = accident_df.append(tmp_df, ignore_index = True)
    if (idx%100==0 or idx==all_idx[-1]):
        #Save after every 100 images
        print('%s-->%s; %d complete'%(full_uri,accident_flag,idx))
        save_obj(accident_df,"full_accident_df.pickle")
        save_obj(all_dict_lbls,"full_all_dict_lbls.pickle")
        upload_blob(bucket_name = "w251-accident-detection",
                    source_file_name = "accident_df.pickle",
                    destination_blob_name = "vision-api-labels/extracted_frames/accident_df.pickle")
        upload_blob(bucket_name = "w251-accident-detection",
                    source_file_name = "all_dict_lbls.pickle",
                    destination_blob_name = "vision-api-labels/extracted_frames/all_dict_lbls.pickle")
        print(datetime.datetime.now())
        print("-----------------------------")
```
