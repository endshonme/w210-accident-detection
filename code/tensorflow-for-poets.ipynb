```python
import pandas as pd
df = pd.read_csv('results_1fps_12Apr19_v2.txt', header = None,sep="|",names = ['path','prob','label_pred','time'])
df.head(2)
```

```python
video_path = []
ttaken = []
mpath = []
for index, row in df.iterrows():
    video_path.append(row['path'].replace('[',''))
    ttaken.append(row['time'].replace(']',''))
df['path'] = video_path
df['time'] = ttaken
df.head()
#print(df.mod_path[0])
#print(df.mod_path[10])
```

```python
import pandas as pd
truth_labels = pd.read_csv('/home/saifrais/w251/final_project/data/video_sample_labels_actuals_1fps.csv')
truth_labels.head(1)
```

```python
mdf = pd.merge(truth_labels[['url','label_act','prob']], df,left_on = 'url', right_on = 'path', how = 'left')
mdf.head(2)
```

```python
from sklearn import metrics
print(metrics.classification_report(mdf['label_act'],mdf['label_pred']))
accuracy = metrics.accuracy_score(mdf['label_act'],mdf['label_pred'])
precision = metrics.precision_score(mdf['label_act'],mdf['label_pred'],average="binary", pos_label="collision")
recall = metrics.recall_score(mdf['label_act'],mdf['label_pred'],average="binary", pos_label="collision")
f1 = metrics.f1_score(mdf.label_act,mdf.label_pred,average="binary", pos_label="collision")
print("Accuracy:{:.3f} Precision:{:.3f} Recall:{:.3f} F1 score:{:.3f}".format(accuracy,precision,recall,f1))
```

```python
filtered_df = mdf[mdf.label_act!=mdf.label_pred]
filtered_df.head(2)
```
